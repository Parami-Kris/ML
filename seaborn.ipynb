```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
dfx=pd.read_csv("c:\parami\paramz\HRDataset_v14.csv")
dfx=dfx[0:25]
dfx=dfx.sort_values(by='EmpID',ascending=True)
dfx=dfx.iloc[:,0:10]
dfx
```

BAR PLOT

```python
plt.figure(figsize=(20,10))
ax=sns.barplot(x=dfx['EmpID'],y=dfx['Salary'],data=dfx)
ax.bar_label(ax.containers[0])
plt.show()
```

LINE AND REL PLOT

```python
sns.relplot(data=dfx,x='EmpID',y='MarriedID',label='MarriedID',kind='scatter')
sns.lineplot(data=dfx,x='EmpID',y='MaritalStatusID',color='red',label='MaritalStatusID')
plt.xlabel('EmpID')
plt.ylabel('Marriage')
plt.show()
```

```python
df=pd.read_csv('c:\parami\paramz\Automobile_data.csv')
plt.figure(figsize=(20,6))
sns.lineplot(data=df['engine-size'],linewidth = 1.5 , label = 'engine size')
sns.lineplot(data=df['wheel-base'],linewidth = 1.5 , label = 'wheel base')
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(x="num-of-cylinders" , y="engine-size" , err_style="bars",data=df)
plt.show()
```

```python
plt.figure(figsize=(14,16))
sns.relplot(x="wheel-base" , y="engine-size" , hue="num-of-cylinders" ,kind='scatter', height=8.5,aspect =1,data=df)
plt.show()
```

CATPLOT

```python
plt.figure(figsize=(11,9))
sns.catplot(x = "city-mpg" , y="body-style" ,kind="violin" ,data=df )
plt.show()
```

```python
plt.figure(figsize=(11,7))
sns.displot(df['city-mpg'], kde=True,bins=20)
plt.show()
```
